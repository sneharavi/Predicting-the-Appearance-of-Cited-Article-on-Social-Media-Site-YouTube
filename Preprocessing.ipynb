```python
import os, json
import numpy as np
import pandas as pd
import glob
```

```python
dict_values={}

PATH = "D:/Fall_2019/Big_Data_Analytics/Altmetric_Data/keys/"

# PATH = "D:/Fall_2019/Big_Data_Analytics/altmetric_clean_sample/"

folder_names = os.listdir(PATH)
columns = ['Altmetric_ID', 'Altmetric_Score', 'Mendeley','CiteULike', 'Twitter', 'Facebook', 'Video',
           'GooglePlus', 'Reddit', 'Blogs', 'Peer_Reviews', 'News', 'F1000', 'Wikipedia', 'Youtube']
df = pd.DataFrame(columns = columns, dtype=int)

count = 0
for folder in folder_names:
    file_names = os.listdir(PATH + folder)

    for file in file_names:            
        with open(os.path.join(PATH, folder, file),encoding="ISO-8859-1") as f:
            data1 = {}
            for line in f:
                data1 = json.loads(line)
                break

            try:
                dict_values['Altmetric_ID']=(data1['altmetric_id'])
            except KeyError:
                dict_values['Altmetric_ID']= np.NaN
                break


            try:
                dict_values['Altmetric_Score']=(data1["altmetric_score"]["score"])
            except KeyError:
                dict_values['Altmetric_Score']=0


            try:
                dict_values['Mendeley']=(data1['counts']['readers']['mendeley'])
            except KeyError:
                dict_values['Mendeley']=0

            try:
                dict_values['CiteULike']=(data1['counts']['readers']['citeulike'])
            except KeyError:
                dict_values['CiteULike']=0

            try:
                dict_values['Twitter']=(data1['counts']['twitter']['posts_count'])
            except:
                dict_values['Twitter']=0

            try:
                dict_values['Facebook']=(data1['counts']['facebook']['posts_count'])
            except:
                dict_values['Facebook']=0

            try:
                dict_values['Video']=(data1['counts']['video']['posts_count'])
            except:
                dict_values['Video']=0

            try:
                dict_values['GooglePlus']=(data1['counts']['googleplus']['posts_count'])
            except:
                dict_values['GooglePlus']=0

            try:
                dict_values['Reddit']=(data1['counts']['reddit']['posts_count'])
            except:
                dict_values['Reddit']=0

            try:
                dict_values['Blogs']=(data1['counts']['blogs']['posts_count'])
            except:
                dict_values['Blogs']=0

            try:
                dict_values['Peer_Reviews']=(data1['counts']['peer_reviews']['posts_count'])
            except:
                dict_values['Peer_Reviews']=0


            try:
                dict_values['News']=(data1['counts']['news']['posts_count'])
            except:
                dict_values['News']=0


            try:
                dict_values['F1000']=(data1['counts']['f1000']['posts_count'])
            except:
                dict_values['F1000']=0

            try:
                dict_values['Wikipedia']=(data1['counts']['wikipedia']['posts_count'])
            except:
                dict_values['Wikipedia']=0


            try:
                if 'youtube' in data1['posts']['video'][0]['url']:
                    dict_values['Youtube']=1
                else:
                    dict_values['Youtube']=0
            except:
                dict_values['Youtube']=0


            if (dict_values['Altmetric_ID'] != np.NaN):
                df.loc[-1] = dict_values  # adding a row
                df.index = df.index + 1  # shifting index
                df = df.sort_index()
                count = count +1



df.to_csv('D:/Fall_2019/Big_Data_Analytics/altmetric_clean_full.csv', index = None, header = True)
print(count)
print(df)
```
